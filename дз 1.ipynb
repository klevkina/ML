    + Сколько женщин было в третьем классе?
      
+ Создайте новый признак __age_group__, который равен:
    + 1, если пассажир был моложе 30 лет;
    + 2, если пассажиру было от 30 до 60 лет;
    + 3, если пассажир был старше 60 лет.

```python
pd.pivot_table(data = df,
               values = 'name',
               index = 'pclass',
               columns = 'sex',
              aggfunc = 'count')
```

```python
n_first_class_men = df.loc[(df['sex']=='male') & (df['pclass']== 1)].shape[0]
n_third_class_women = df.loc[(df['sex']=='female') & (df['pclass']== 3)].shape[0]


print(f'Количество мужчин в первом классе: {n_first_class_men}')
print(f'Количество женщин в третьем классе: {n_third_class_women}')
```

```python
def age_group(age):
    if age < 30: 
        return 1
    elif age >= 30 and age <= 60:
        return 2
    elif age > 60:
        return 3

df['age_group'] = df['age'].apply(age_group)
df.head()
```

**Задание 3**\
Найдите самое популярное имя среди пассажиров:
+ мужчин;
+ женщин.

```python
status_list = df['name'].str.split().str[1].unique()
def split_cond(name):
    if 'Mrs'in name and '(' in name and ')' in name:
        return name.split('(')[1].split(')')[0].strip()
    else:
        for status in status_list:
            if status in name:
                return name.split(status)[1].strip().split(' ')[0]
    return name.strip().split()[0]
```

```python
df['name_only'] = df['name'].apply(split_cond)
df.groupby('sex')['name_only'].value_counts().groupby(level=0).head(1)
```
